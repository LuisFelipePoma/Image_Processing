```python
import numpy as np
import skimage #scikit-image
import matplotlib.pyplot as plt
```

```python
im = skimage.data.coins()
plt.imshow(im, cmap='gray')
```

```python
im
```

```python
#@title Cambiar valor de t { run: "auto" }
t = 132 #@param {type:"slider", min:0, max:255, step:1}
im_th = im.copy()
im_th[im_th<t] = 0
im_th[im_th>=t] = 255
plt.figure()
plt.imshow(im_th, cmap='gray')
```
