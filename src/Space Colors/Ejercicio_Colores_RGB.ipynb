```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#@title Colores RGB { run: "auto" }
R = 255 #@param {type:"slider", min:0, max:255, step:1}
G = 255 #@param {type:"slider", min:0, max:255, step:1}
B = 255 #@param {type:"slider", min:0, max:255, step:1}

#vamos a crear una imagen de 100x100x3 (la tercera dimensión son los colores)
imagen = np.zeros((100,100,3), dtype=np.uint8)
imagen[:,:,0] = R
imagen[:,:,1] = G
imagen[:,:,2] = B
plt.figure(0)
plt.title( "({}, {}, {})".format(R,G,B))
plt.imshow(imagen)
plt.show()
```
